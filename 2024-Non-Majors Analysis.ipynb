### Renaming and Selecting Relevant Columns

1. **Merge multiple renaming dictionaries** dict_set_one, dict_set_two, etc into a single dictionary called renaming_dict
2. **Rename dataset columns** using df.rename() to replace long column names with more concise and readable ones.
3. **Select only relevant columns**:
   - I extracted the values from renaming_dict (i.e., the new column names).
   - Then, I filtered the DataFrame to keep only columns that exist in df.columns after renaming.

```python
renaming_dict = {**dict_set_one, **dict_set_two, **dict_set_three, **dict_set_four}
```

```python
df.rename(columns=renaming_dict, inplace=True)
```

```python
relevant_columns = [col for col in renaming_dict.values() if col in df.columns]
df = df[relevant_columns]
```

### Standardizing Major and Race Categories

To simplify analysis, I consolidated **similar categories** in the Major and Race columns:

1. **Majors Consolidation:**
   - Engineering-related fields (**Mechanical, Electrical, Civil Engineering**) → **"Engineering"**  
   - Computing-related fields (**Computer Science, IT, Software Engineering, Data Science**) → **"Computing"**  
   - Business-related fields (**Business Administration, Finance, Marketing**) → **"Business"**  
   - Science fields (**Biology, Chemistry, Physics**) → **"Science"**  

2. **Race Standardization:**
   - Converts multi-word categories to simpler labels (e.g., **"White/Caucasian" → "White"**).
   - Groups **"Two or More Races"** as **"Multiracial"**.

```python
if "major" in org_df .columns:
    df["major"] = df["major"].replace({
        "Mechanical Engineering": "Engineering",
        "Electrical Engineering": "Engineering",
        "Civil Engineering": "Engineering",
        "Computer Science": "Computing",
        "Information Technology": "Computing",
        "Software Engineering": "Computing",
        "Data Science": "Computing",
        "Business Administration": "Business",
        "Finance": "Business",
        "Marketing": "Business",
        "Biology": "Science",
        "Chemistry": "Science",
        "Physics": "Science",
        "Other": "Other"
    })
```

```python
if "race" in org_df .columns:
    df["race"] = df["race"].replace({
        "White/Caucasian": "White",
        "Black or African American": "Black",
        "Asian": "Asian",
        "Hispanic/Latino": "Hispanic",
        "Two or More Races": "Multiracial",
        "Other": "Other"
    })    
```

#### Drop irrelevant columns: 'Timestamp' is not needed)


```python
if 'Timestamp' in df.columns:
    df = df.drop(columns=['Timestamp'])
```

### Which computing courses are students most interested in taking next?

```python
computing_interest_cols = [
    'web_development',
    'cybersecurity',
    'app_development',
    'data_analytics',
    'machine_learning_ai',
    'computer_programming',
    'game_design',
    'hardware_installation_repair'
]
```

```python
for col in computing_interest_cols:
    df[col] = df[col].map({'Yes': 1, 'No': 0}).fillna(0)


course_interest_counts = df[computing_interest_cols].sum().sort_values(ascending=True)
```

```python
plt.figure(figsize=(10, 6))
plt.barh(course_interest_counts.index.str.replace('_', ' ').str.title(), course_interest_counts.values, color='teal', edgecolor='black')
plt.xlabel("Number of Interested Students", fontsize=12)
plt.title("Which Computing Courses Are Students Most Interested In?", fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```
